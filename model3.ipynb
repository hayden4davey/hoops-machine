```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
bb_df = pd.read_csv("Resources/CleanHoopsData.csv")
bb_df.head()
```

```python
bb_df['Pos'] = bb_df['Pos'].map({'PG':1, 'SG':2, 'SF':3, 'PF':4,'C':5})
```

```python
bb_df['BMI'] = bb_df['Weight'] / ((bb_df['Height']/100)*(bb_df['Height']/100))
```

```python
target = bb_df['Pos'].values.reshape(-1,1)
target_names = ['PG','SG','SF','PF','C']
```

```python
player_names = bb_df['Player']
```

```python
data = bb_df.drop(['Tm','Player','Pos', 'Height','Weight'], axis=1)
feature_names = data.columns
data.head()
```

```python
list(data.columns)
```

```python
data['Year'] = data['Year'] - 1978
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train.ravel())
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
