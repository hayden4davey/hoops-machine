```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
bb_df = pd.read_csv("Resources/CleanHoopsData.csv")
bb_df.head()
```

```python
bb_df['Pos'] = bb_df['Pos'].map({'PG':1, 'SG':2, 'SF':3, 'PF':4,'C':5})
```

```python
plt.scatter(bb_df['Pos'], bb_df['Height'], edgecolors='black', alpha=0.5)
plt.grid(axis='both', which='major')
plt.title('Position Played by Height')
plt.xlabel('Position')
plt.ylabel('Height(cm)')
plt.xticks(np.arange(0, 6, step=1))
```

```python
plt.savefig('static/images/HeightVis.png')
```

```python
bb_df['BMI'] = bb_df['Weight'] / ((bb_df['Height']/100)*(bb_df['Height']/100))
```

```python
plt.scatter(bb_df['Pos'], bb_df['BMI'], edgecolors='black', alpha=0.5)
plt.grid(axis='both', which='major')
plt.title('Position Played by BMI')
plt.xlabel('Position')
plt.ylabel('Body Mass Index')
plt.xticks(np.arange(0, 6, step=1))
```

```python
plt.savefig('static/images/BMIVis.png')
```

```python
plt.scatter(bb_df['Pos'], bb_df['AST%'], edgecolors='black', alpha=0.5)
plt.grid(axis='both', which='major')
plt.title('Assist Percentage by Player Position')
plt.xlabel('Position')
plt.ylabel('Assist Percentage')
plt.xticks(np.arange(0, 6, step=1))
```

```python
plt.savefig('static/images/AstPerVis.png')
```

```python
plt.scatter(bb_df['Pos'], bb_df['AST'], edgecolors='black', alpha=0.5)
plt.grid(axis='both', which='major')
plt.title('Assists by Player Position')
plt.xlabel('Position')
plt.ylabel('Assists per Game')
plt.xticks(np.arange(0, 6, step=1))
```

```python
plt.savefig('static/images/AstVis.png')
```
