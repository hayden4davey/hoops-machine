```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
import sklearn.datasets
```

```python
data = pd.read_csv("Resources/CleanHoopsData.csv")
data.head()
```

```python
data = data.drop(['Tm','Player'], axis=1)
```

```python
data['Year'] = data['Year'] - 1978
data.head()
```

```python
X = data.drop("Pos", axis=1)
y = data["Pos"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1, stratify=y)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Point guards 

```python
encoded_y_train = y_train.map({'PG':1, 'SG':0, 'SF':0, 'PF':0,'C':0})
encoded_y_test = y_test.map({'PG':1, 'SG':0, 'SF':0, 'PF':0,'C':0})
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=40))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    encoded_y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, encoded_y_test, verbose=2)
print(
    f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model.save("pg_model.h5")
```
